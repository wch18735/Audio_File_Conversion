### GET SAMPLE DATA LIST

```python
import os

pcm_sample_list = os.listdir(SAMPLE_PCM_PATH)
wav_sample_list = [pcm_name.split('.')[0]+'.wav' for pcm_name in pcm_sample_list]
```

### DEFINE Pcm2Wav CLASS

```python
import wave

class Pcm2Wav(object):
    def __init__(self):
        self.sample_rate = 16000
        self.channels = 1
        self.bits = 16

    def __call__(self, pcm_path, wav_path):
        pcm = None

        with open(pcm_path, 'rb') as pcmdata:
            pcm = pcmdata.read()

        with wave.open(wav_path, 'wb') as wavdata:
            wavdata.setnchannels(self.channels)
            wavdata.setsampwidth(self.bits // 8)
            wavdata.setframerate(self.sample_rate)
            wavdata.writeframes(pcm)
```

```python
pcm2wav = Pcm2Wav()
```

```python
# get file name concatenated with file path

pcm_file = os.path.join(SAMPLE_PCM_PATH, pcm_sample_list[0])
wav_file = os.path.join(CONVERT_WAV_PATH, wav_sample_list[0])
```

```python
# Do convert
pcm2wav(pcm_file, wav_file)
```
